# Initial Loading and Exploration

## Audio Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load audio metadata into data frame
audio_metadata = pd.read_csv("data/LJSpeech-1.1/metadata.csv")
print("First 5 metadata samples:")
display(audio_metadata.head())

# Show waveform of first 3 samples:
for file_name in audio_metadata.head(3)["id"]:

    # Load WAV file
    sample_rate, audio_data = wavfile.read(f"data/LJSpeech-1.1/wavs/{file_name}.wav")

    # Create time axis in seconds
    time = [i / sample_rate for i in range(len(audio_data))]

    # Plot waveform of sample
    plt.figure(figsize=(12, 4))
    plt.plot(time, audio_data)
    plt.title(f"Waveform of {file_name}")
    plt.xlabel("Time (sec)")
    plt.ylabel("Amplitude")
    plt.show()
    
    # Show transcript
    transcript = audio_metadata[audio_metadata["id"] == file_name]["sentence"].values.item()
    print(f"Transcription: {transcript}")
```

## Screen Annotation Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load training CSV into data frame
train_df = pd.read_csv("data/screen_annotation/train.csv")

print(f"Number of training photos: {len(train_df)}")
print(f"First 5 rows of training dataset:")
display(train_df.head())

# Count UI elements per screen
def count_elements(annotation_str, UI_type="BUTTON"):
    if pd.isna(annotation_str):
        return 0
    return annotation_str.count(UI_type)

for UI_type in ["BUTTON", "TEXT", "PICTOGRAM"]:

    train_df["num_elements"] = train_df["screen_annotation"].apply(count_elements, args=(UI_type,))

    # Histogram of number of UI elements per screen
    train_df["num_elements"].hist(bins=30)
    plt.xlabel(f"Number of {UI_type} Elements")
    plt.ylabel("Number of Screens")
    plt.title(f"Distribution of {UI_type} Elements (Training Data)")
    plt.show()

    # Print first annotation string
    print(f"First screen annotation:\n{train_df.iloc[0]['screen_annotation']}")
```
